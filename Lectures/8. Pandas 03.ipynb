## 3. Extra Work

```python
start = dt.datetime(2023,1,2)
end = dt.date.today()

yf.pdr_override()
price_AAPL = data.DataReader('AAPL', start, end)
```

```python
price_AAPL.head()
```

```python
price_AAPL['daily_diffrence'] = price_AAPL['Adj Close'].diff()
price_AAPL.head()
```

```python
price_AAPL['signal'] = np.where(price_AAPL['daily_diffrence'] > 0, 1, 0)
price_AAPL.head()
```

```python
price_AAPL['position'] = price_AAPL['signal'].diff()
price_AAPL.head(10)
```

```python
price_AAPL[price_AAPL['position']==1].index
```

```python
price_AAPL.loc[price_AAPL['position']==1, 'Adj Close']
```

```python
plt.figure(figsize=(15,8))

plt.plot(price_AAPL['Adj Close'], c = 'r', lw = 2)

plt.plot(price_AAPL[price_AAPL['position']==1].index, 
         price_AAPL.loc[price_AAPL['position']==1, 'Adj Close'], 'o', c = 'g', markersize = 7)

plt.plot(price_AAPL[price_AAPL['position']==-1].index, 
         price_AAPL.loc[price_AAPL['position']==-1, 'Adj Close'], 'x', c = 'k', markersize = 7)


plt.grid()
plt.show()
```

## 5.1

```python
# pip install wbdata
```

```python
import wbdata as wb
```

```python
countries = ['USA', 'CHN']

data = wb.get_dataframe({"NY.GDP.PCAP.PP.CD":'GDP_Per_Capita', "SP.DYN.LE00.IN": "life_Expectancy"},
                        data_date=(pd.to_datetime("2010"), pd.to_datetime('2020')),
                       country=countries, convert_date=False)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data['log(gdp per capita)'] = np.log(data["GDP_Per_Capita"])
```

```python
#data
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(data['log(gdp per capita)'])
y = data['life_Expectancy']

ols_model = sm.OLS(y, X)
```

```python
results = ols_model.fit()
```

```python
print(results.summary())
```
